```python
import tensorflow as tf
from tensorflow import keras

```

```python
from keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
img = image.load_img(r'/content/c7.jpeg',target_size=(200,200))
```

```python
img
```

```python
type(img) # It is not a numpy arrray
```

```python
datagen = ImageDataGenerator(
    rotation_range = 30,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip = True,
    brightness_range = (0.5,1.5),
    width_shift_range = 0.2,
    height_shift_range = 0.2,
    fill_mode = 'reflect'
)
```

```python







img = image.img_to_array(img) #converting img into Array
```

```python
type(img)
```

```python
img.shape
```

```python
input_batch = img.reshape(1,200,200,3) # Adding one extra dimension to tell that we are going to pocess 1 number of images
```

```python
input_batch.shape
```

```python
i =0
for output in datagen.flow(input_batch,batch_size=1,save_to_dir =r'/content/sample_data/aug'): # needed to speicfy a directory to save generated images
  i+=1
  if i == 10:
    break
```
